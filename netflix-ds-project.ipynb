```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the data from the csv file
netflix_df = pd.read_csv('netflix_data.csv')
```

```python
# Filtering the data to remove TV shows and store as netflix_subset
netflix_subset = netflix_df[netflix_df['type'] == 'Movie']
```

```python
# Creating subset of movie with the columns : 'title', 'country', 'genre', 'release_year', 'duration'
netflix_movies = netflix_subset[['title', 'country', 'genre', 'release_year', 'duration']]
```

```python
# Filtering out short movies that are strictly less than 60 minutes
short_movies = netflix_movies[netflix_movies['duration'] < 60]
```

```python
# Create ab empty list to store colors
colors = []
```

```python
# Iterate over rows of netflix_movies and assign color based on genre
for index, row in netflix_movies.iterrows():
    if row['genre'] == 'Children':
        colors.append('blue')
    elif row['genre'] == 'Documentaries':
        colors.append('green')
    elif row['genre'] == 'Stand-Up':
        colors.append('red')
    else:
        colors.append('black')
```

```python
# Initialize the figure and create a scatter plot
fig = plt.figure(figsize=(12,8))
plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors, alpha=0.6)

# Create a title and labels
plt.title('Movie duration by year of release')
plt.xlabel('Release year')
plt.ylabel('Duration (min)')

# Show the plot
plt.show()
```
